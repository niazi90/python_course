Write a program that takes salary as input. Using conditional statements, calculate the final tax
based on the following rules:
• If the salary is less than 30,000 → Tax rate is 5%
• If the salary is between 30,000 and 70,000 → Tax rate is 15%
• If the salary is greater than 70,000 → Tax rate is 25%


```python
salary=float(input("Enter your salary:"))
if salary<30000:
    
    tax=salary * 5/100 
    print(" tax apply is 5%")
elif salary>=30000 and salary<=70000:
    tax= salary * 15/100
    print(" tax apply is 15%")
elif salary>70000:
    tax=salary * 25/100
    print(" tax apply is 25%")
else:
    print("Invalid input")
print('salary:',salary)
print('Tax:',tax)
print('final salary:',salary-tax)
```

Given a list of words: words = ["apple", "banana", "kiwi", "cherry", "mango"] 
Create a dictionary that maps each word to its corresponding length.
Example Output: ({"apple": 5, "banana": 6, "kiwi": 4, "cherry": 6, "mango": 5})

```python
word=["apple","banana","kiwi","cherry","mango"]
dword={}
for i in word:
    dword[i]=len(i)
print(dword)


```

Task 3:
Create a Chat System using Object-Oriented Programming (OOP) concepts.
You need to create the following classes:
• User
• Message
• ChatRoom
The system should implement the following functionalities:
• Sending messages
• Viewing chat history
• User joining and leaving the chat room


```python
class User:
    def __init__(self,name):
        self.name=name
    def send_message(self,chatroom,message):
        message=Message(self.name,message)
        chatroom.add_message(message)



class Message:
    def __init__(self,sender,text):
        self.sender=sender
        self.text=text

class ChatRoom:
    def __init__(self):
        self.users = []
        self.messages = []

    def join(self, user):
        self.users.append(user)
        print(user.name, "joined the chat room.")

    def leave(self, user):
        self.users.remove(user)
        print(user.name, "left the chat room.")

    def add_message(self, message):
        self.messages.append(message)

    def show_history(self):
        print("\nChat History:")
        for msg in self.messages:
            print(msg.sender + ":", msg.text)


# Main Program
room = ChatRoom()

user1 = User("Ali")
user2 = User("Sara")

room.join(user1)
room.join(user2)

user1.send_message(room, "Hello everyone!")
user2.send_message(room, "Hi Ali!")

room.show_history()

room.leave(user2)
```
