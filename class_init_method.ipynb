```python
class Dog:
    def __init__(self):
        self.name = "Buddy"
    def bark(self):
        print(f"{self.name} says: Woof!")
dog1=Dog()
dog1.bark()
```

```python
class Car:
    def __init__(self,brand,color):
        self.brand=brand
        self.color=color
car1=Car("Toyota","Red")
print(car1.brand)
print(car1.color)
car2=Car("civic","white")
print(car2.brand)
print(car2.color)
        
```
