```python
import numpy as np
import time
```

```python
arr=np.array([[1,2,3],[5,7,9]])
print(arr.shape)
print(arr.ndim)
print(arr.dtype)
```

```python
py_list=list(range(1_000_000))
py_result=[]
start_time=time.time()
for i in py_list:
    py_result.append(i+5)
end_time=time.time()
print('Python List Time',end_time-start_time)
```

```python
np_array=np.arange(1_000_000)
start_time=time.time()
np_result=np_array+5
end_time=time.time()
print('Numpay Array Time',end_time-start_time)
```
