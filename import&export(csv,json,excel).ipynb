Key concept
Reading CSV: pd.read_csv()

1:   parse_dates=['column']->convert date columns
2:    usecols=[...]-> load specific columns
3:    nrows=100    -> load subset to save memory  

Reading Excel: pd.read_excel('file.xlsx',sheet_name=['Sheet'])

Reading JSON: pd.read_json('file.json')     



Key Concepts

    Exporting DataFrames:
                    df.to_csv('file.csv',index=False)

                    df.to_excel('file.xlsx',index=False)

                    df.to_json('file.json',orient='records')

key concepts
     1:   Memory Tips:

            Use dtype to specify column types
     2:     Use parse_dates when working     with date/time data
     3:     Expoting:
            CSV,Excel,JSON  -> easy sharing & interoperability
     4: For large datasets, load only needed columns or rows

```python
import pandas as pd
import openpyxl

df = pd.read_csv("retail_data.csv",parse_dates=["Date"],dtype={
    "Category":"category",
    "Region":"category"
})
print(df.info())
```

```python
clean_df=df[df["Sales"]>0]
subset=clean_df[["Date","Catagoery","Sales","Regin"]]

```

```python
subset.to_excel("subset_sales.xlsx", sheet_name="sales")
subset.to_json("subset_sales.json",
               orient="records",
               date_format="iso")
print("File exported successfully")
subset.info()
```
