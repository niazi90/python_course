```python
import pandas as pd
```

```python
data={
    'Name':['Azhar','ALi','Akram','Asif','Ahsan'],
    'Age':[23,24,22,25,21],
    'city':['Karachi','Lahore','Islamabad','Peshawar','Quetta'],
    'Salary':[50000,60000,55000,65000,45000]
}
df=pd.DataFrame(data)
print(df)
```

```python
subset_loc=df.loc[1:3,['Name','city']]
print("Subset with loc:\n",subset_loc)
```

Work with indexing

```python
subset_iloc=df.iloc[0:2,1:3]
print("Subset with iloc:\n",subset_iloc)
```

filtering

```python
high_salary=df[df['Salary']>55000]
print("Employees with Salary > 55000:\n",high_salary)
```

update the salary 

```python
df.loc[df['Age']<25 ,'Salary']=55000
print("Low Age Salary change:\n",df.loc[df['Age']< 25,'Salary'])
```
