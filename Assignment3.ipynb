```python
# Qno1 Make a calculator using python with addition subtraction multiplication division and power
print("calculator app")
num1=int(input("Enter Your First number: "))
num2=int(input("Enter Your Second number: "))
operator=input("Which Operator you use +,-,*,/,%,**: ")
if operator=="+":
    print(f" Addition of {num1} and {num2} are = {num1+num2}")
elif operator=="-":
       print(f" Subtraction of {num1} and {num2} are = {num1-num2}")
elif operator=="*":
       print(f" Multiplication of {num1} and {num2} are = {num1*num2}")
elif operator=="/":
      print(f" Division of {num1} and {num2} are = {num1/num2}")
elif operator=="%":
       print(f" Modulus of {num1} and {num2} are = {num1%num2}")
elif operator=='**':
       print(f" Power of {num1} and {num2} are = {num1**num2}")
    
else:
    print("Enter Your Correct Operator")
```

```python
# Qno2  write a program to check if there is any numeric value in the list using for loop
a = ["Apple", "karachi", 90, 67, 55, 4, 34, "lahore"]
found = False
for i in a:
    if isinstance(i, (int, float)):
        found = True
        break
if found:
    print("Yes, list contains numeric value")
else:
    print("No numeric value in list")
```

```python
# Qno4 write a python program to sum all the numeric items in a dictionary
a = {
    "x": 10,
    "y": 20,
    "z": "Apple",
    "w": 30
}
total = 0
for value in a.values():
    if isinstance(value, (int, float)):
        total += value

print("Sum of numeric values:", total)
```

```python
# Qno5 write a program to identify a duplicate value from list
a = [1, 2, 3, 4, 2, 5, 6, 3, 7,9,9]
duplicates = []
for i in a:
    if a.count(i) > 1 and i not in duplicates:
        duplicates.append(i)

print("Duplicate values:", duplicates)
```
