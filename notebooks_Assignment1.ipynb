```python
# Qno 7 write a program that take 5 input from the user  for different subject's marks,total it
# and generate marks sheet using Grade
Math=int(input("Enter Your Math Number:"))
Urdu=int(input("Enter Your Urdu Number:"))
Computer=int(input("Enter Your Computer Number:"))
physic=int(input("Enter Your physic Number:"))
Chemistry=int(input("Enter Your Chemistry Number:"))

total=Math+Urdu+Computer+physic+Chemistry

print(f"Total marks are: {total} out of 500")
grade=(total/500)*100

if(grade<=100 and total<=500):
    if(grade<=100 and grade>=90):
        print(f"Your Math:{Math}, Urdu:{Urdu},Computer:{Computer},Physic:{physic},Chemistry:{Chemistry} marks which is total:{total} marks and percetage {grade} is: You Got  A+")
    elif(grade<=90 and grade>=80):
       print(f"Your Math:{Math}, Urdu:{Urdu},Computer:{Computer},Physic:{physic},Chemistry:{Chemistry} marks which is total:{total} marks and percetage {grade} is: You Got  A")
    elif(grade<=80 and grade>=70):
        print(f"Your Math:{Math}, Urdu:{Urdu},Computer:{Computer},Physic:{physic},Chemistry:{Chemistry} marks which is total:{total} marks and percetage {grade} is: You Got  B+")
    elif(grade<=70 and grade>=60):
         print(f"Your Math:{Math}, Urdu:{Urdu},Computer:{Computer},Physic:{physic},Chemistry:{Chemistry} marks which is total:{total} marks and percetage {grade} is: You Got  B")
    elif(grade<=60 and grade>=50):
         print(f"Your Math:{Math}, Urdu:{Urdu},Computer:{Computer},Physic:{physic},Chemistry:{Chemistry} marks which is total:{total} marks and percetage {grade} is: You Got  C")
    elif(grade<=50 and grade>=40):
         print(f"Your  Math:{Math}, Urdu:{Urdu},Computer:{Computer},Physic:{physic},Chemistry:{Chemistry} marks which is total:{total} marks and percetage {grade} is: You Got  D")
    elif(grade<=40 and grade>=30):
         print(f"Your Math:{Math}, Urdu:{Urdu},Computer:{Computer},Physic:{physic},Chemistry:{Chemistry} marks which is total:{total} marks and percetage {grade} is: You Got  E")
    else:
        print(f"Your  total {total} marks and  percetage {grade} is: You are Fail")
else:
    print("Enter Your Correct entry")





```

```python
# Qno 9 calculator
print("calculator app")
num1=int(input("Enter Your First number: "))
num2=int(input("Enter Your Second number: "))
operator=input("Which Operator you use +,-,*,/,% : ")
if operator=="+":
    print(f" Addition of {num1} and {num2} are = {num1+num2}")
elif operator=="-":
       print(f" Subtraction of {num1} and {num2} are = {num1-num2}")
elif operator=="*":
       print(f" Multiplication of {num1} and {num2} are = {num1*num2}")
elif operator=="/":
      print(f" Division of {num1} and {num2} are = {num1/num2}")
elif operator=="%":
       print(f" Modulus of {num1} and {num2} are = {num1%num2}")
else:
    print("Enter Your Correct Operator")
```
